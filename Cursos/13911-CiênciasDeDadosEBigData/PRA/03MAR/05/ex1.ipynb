```python
import pandas as pd
import numpy
```

```python
crimes = pd.read_csv('crimes.csv')
entrevistas = pd.read_csv('entrevistas.csv')
pessoas = pd.read_csv('pessoas.csv')
veiculos = pd.read_csv('veiculos.csv')
associacoes = pd.read_csv('associacoes.csv')
```

```python
suspeitos_altura = pessoas[pessoas['altura'] >= 195]
```

```python
ponte_preta = associacoes[associacoes['clube'] == 'Associação Atlética Ponte Preta']
hipica = associacoes[associacoes['clube'] == 'Sociedade Hípica de Campinas']
```

```python
clubes_cruzados = pd.merge(ponte_preta, hipica, on='nome', suffixes=('_ponte', '_hipica'))
```

```python
veiculos_suspeitos = veiculos[
    (veiculos['marca'] == 'Honda') & 
    (veiculos['cor'] == 'Azul Claro') & 
    (veiculos['placa'].str.contains('33') | veiculos['placa'].str.endswith('89'))
]
```

```python
suspeitos_finais = suspeitos_altura.merge(clubes_cruzados, on='nome')
```

```python
assassino = suspeitos_finais.merge(veiculos_suspeitos, on='nome')
```
