```python
import pandas
import numpy
```

```python
elections = pandas.read_csv('data/elections.csv')
```

```python
print(elections.columns)
```

```python
elections.iloc[0, 1]
```

```python
elections[((elections['Year'] == 2024) | (elections['Year'] == 2020))]
```

```python
elections.groupby('Candidate').agg('max')
# Agrupa por candidato e mostra o ano mais recente e o partido mais recente de cada um
```

```python
elections[['Candidate', 'Year', 'Party']].groupby('Candidate').agg('max')
# Agrupa por candidato e mostra o ano mais recente e o partido mais recente de cada um, mas só mostra as colunas de candidato, ano e partido
```

```python
elections.groupby('Candidate').count()
# Quantidade de vezez que cada candidato apareceu
```

```python
elections.groupby('Candidate').value_counts()
# Mostra os dados agruoados por candidato e por ano, partido, etc
```

```python
elections.groupby('Year').filter(lambda x: x['%'].max() > 45)
# Filtra os anos em que algum candidato teve mais de 45% dos votos
```

```python
elections.groupby('Year').filter(lambda x: x['%'].max() > 45)
# Filtra os anos em que algum candidato teve mais de 45% dos votos, mas só mostra os dados desses anos
```

```python
elections.sort_values('%', ascending=False)
# Ordena os dados por porcentagem de votos, do maior para o menor
```

```python
elections.sort_values('%', ascending=False).groupby('Party').first()
# Mostra o partido mais votado de cada partido, ou seja, o candidato mais votado de cada partido
```

```python
elections[['Candidate', 'Party']].groupby('Party').agg('count').sort_values('Candidate', ascending=False).head(5)
# Mostra a quantidade de candidatos de cada partido, ordenado do maior para o menor, e mostra só os 5 primeiros
```
