### Questão 11
Qual é a data exata que apresentou a maior proporção de casos críticos neste conjunto de dados, exigindo que você crie uma nova coluna chamada `critical_ratio` resultante da divisão dos pacientes em unidades de terapia intensiva pelo total de pacientes hospitalizados, descartando a coluna original de pacientes hospitalizados logo em seguida para avaliar a sobrecarga aguda do sistema de saúde?

```python
dataset['critical_ratio'] = dataset['icu_patients'] / dataset['hosp_patients']
dataset.drop(columns = ['hosp_patients'])
dataset.sort_values('critical_ratio', ascending=False)['date'].iloc[0]
```

### Questão 12
Quais são os países que mantiveram uma média diária de novos casos superior a dez mil ao longo de todo o seu histórico registrado neste conjunto de dados, para que possamos identificar os surtos mais descontrolados?

```python
dataset.groupby('location').filter(lambda row : row['new_cases'].mean() > 10000)['location'].unique()
# groupby('location') -> agrupa por localização
# filter(lambda row : row['new_cases'].mean() > 10000) -> filtra os grupos onde a média de new_cases é maior que 10000
# ['location'].unique() -> retorna as localizações únicas dos grupos filtrados
```

### Questão 13
Quais países registraram um desvio padrão superior a mil unidades na métrica de pacientes internados em unidades de terapia intensiva considerando todo o período documentado neste conjunto de dados, de forma a mapear o caos hospitalar?

```python
dataset.groupby('location').filter(lambda row : row['icu_patients'].std() > 1000)['location'].unique()
```

### Questão 14
Qual é o maior valor de prevalência de diabetes registrado apenas para os países onde a taxa de mortalidade ultrapassou dois por cento da população neste conjunto de dados, exigindo a criação prévia de uma coluna chamada `death_population_ratio` resultante do total acumulado de mortes dividido pela população para analisar fatores de risco em cenários extremos?

```python
dataset['death_population_ratio'] = dataset['total_deaths'] / dataset['population']
dataset.groupby('location').filter(lambda row : row['death_population_ratio'].max() > 0.02)['diabetes_prevalence'].max()
```

### Questão 15
Quais países conseguiram manter uma mediana diária de novos testes superior a cinco mil ao longo de toda a sua trajetória neste conjunto de dados, de forma a podermos premiar as nações com rastreamento populacional constante?

```python
dataset.groupby('location').filter(lambda row : row['new_tests'].median() > 5000)['location'].unique()
```
