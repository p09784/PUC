### `Series`

`Series` representa uma coluna de um `DataFrame` ou, de forma mais geral, pode ser qualquer objeto semelhante a um vetor unidimensional. Ela contém tanto uma sequência de valores do mesmo tipo quanto uma sequência de rótulos de dados chamada índice.

```python
series = pandas.Series(['Welcome', 'to', 'Data Science', '!'])
```

```python
series.values
```

```python
series.index
```

O `index` de uma `Series` é, normalmente, uma lista sequencial de números inteiros começando em 0. Opcionalmente, uma lista de índices desejados especificada manualmente pode ser passada para o argumento `index`.

```python
series = pandas.Series([-1, 10, 2], index = ['a', 'b', 'c'])
```

```python
series.values
```

```python
series.index
```

Os índices também podem ser alterados após a inicialização.

```python
series.index = ['first', 'second', 'third']
```

```python
series.index
```

#### Seleção

É possível selecionar um único valor ou um conjunto de valores de uma `Series` e, para isso, existem três métodos principais: um único rótulo, uma lista de rótulos e uma condição de filtragem.

```python
series = pandas.Series([4, -2, 0, 6], index = ['a', 'b', 'c', 'd'])
```

##### Seleção usando um único rótulo

```python
# retorna o valor armazenado no rótulo de índice 'a'
series['a']
```

##### Seleção usando uma lista de rótulos

```python
# retorna uma Series com os valores armazenados nos rótulos de índice 'a' e 'c'
series[['a', 'c']]
```

##### Seleção usando uma condição de filtragem

```python
# retorna uma Series indicando quais elementos satisfazem a condição
series > 0
```

```python
# retorna apenas os elementos que satisfazem a condição
series[series > 0]
```

### `DataFrame`

#### Criando um `DataFrame` a partir do dicionário

```python
fruits = pandas.DataFrame([{'Fruit' : 'Kiwi', 'Price' : 5.49}, {'Fruit' : 'Orange', 'Price' : 3.99}])
```

```python
fruits = pandas.DataFrame({'Fruit' : ['Kiwi', 'Orange'], 'Price' : [5.49, 3.99]})
```

#### Criando um `DataFrame` a partir da lista de listas

```python
fruits = pandas.DataFrame([['Kiwi', 5.49], ['Orange', 3.99]], columns = ['Fruit', 'Price'])
```

#### Criando um `DataFrame` a partir de uma `Series`

```python
fruit = pandas.Series(['Kiwi', 'Orange'], index = [1, 2])
```

```python
price = pandas.Series([5.49, 3.99], index = [1, 2])
```

```python
fruits = pandas.DataFrame({'Fruit' : fruit, 'Price' : price})
```

#### Extraindo dados de um `DataFrame`

Existem várias maneiras de extrair linhas e colunas de interesse de um `DataFrame`, como obter as primeiras ou últimas n linhas, selecionar dados com um determinado rótulo ou acessar dados em uma posição específica, e todas essas três abordagens se mostram úteis em diferentes contextos.

```python
# extrai as 5 primeiras linhas do DataFrame
elections.head()
```

```python
# extrai as 5 últimas linhas do DataFrame
elections.tail()
```

##### Extração baseada em rótulos usando o `.loc`

`.loc` utiliza dois argumentos, um para os rótulos das linhas e outro para os rótulos das colunas. Cada argumento do `.loc` pode ser um único valor, uma lista, um slice ou uma máscara booleana. `.loc` seleciona itens pelos rótulos de linhas e colunas.

```python
elections = pandas.read_csv('data/elections.csv')
```

```python
elections.loc[0, 'Candidate']
```

```python
elections.loc[[87, 25, 179], ['Year', 'Party', '%']]
```

```python
elections.loc[[87, 25, 179], 'Popular vote':'%']
```

```python
elections.loc[:, ['Year', 'Party', '%']]
```

```python
elections.loc[[87, 25, 179], :]
```

```python
elections.loc[[87, 25, 179], 'Popular vote']
```

```python
# observe que, se passarmos 'Popular vote' em uma lista, o resultado será um DataFrame
elections.loc[[87, 25, 179], ['Popular vote']]
```

```python
elections.loc[elections['Year'] == 2008, ['Year', 'Candidate']]
```

```python
# se usarmos apenas um argumento no .loc, o pandas o aplica às linhas e retorna todas as colunas
elections.loc[[180, 181]]
```

##### Extração baseada em inteiros usando o `.iloc`

`.iloc` seleciona itens pela posição inteira das linhas e colunas. Cada argumento do `.iloc` pode ser um único valor, uma lista ou um slice. `.iloc` seleciona itens pelas posições inteiras de linhas e colunas.

```python
elections.iloc[0, 1]
```

```python
elections.iloc[[1, 2, 3], 1]
```

```python
elections.iloc[[1, 2, 3], [0, 1, 2]]
```

```python
elections.iloc[[1, 2, 3], 0:3]
```

```python
elections.iloc[:, 0:3]
```

```python
elections.iloc[138:144]
```

##### Extração dependente do contexto usando `[]`

O operador de seleção `[]` é o mais confuso de todos, mas também o mais utilizado, aceitando apenas um único argumento, que pode ser um slice de números de linha, uma lista de rótulos de coluna ou um único rótulo de coluna, ou seja, o `[]` depende do contexto.

```python
elections['Candidate']
```

```python
elections[['Year', 'Candidate', 'Result']]
```

```python
elections[3:7]
```

```python
elections[elections['Popular vote'] > 60000000]
```

##### Operadores booleanos

Filtrar com múltiplas condições é feito combinando máscaras booleanas usando **comparações bit a bit**.

| Símbolo | Uso       | Significado                     |
|---------|-----------|--------------------------------|
| `~`     | `~p`      | Retorna a negação de `p`       |
| `\|`    | `p \| q`  | `p` OU `q`                     |
| `&`     | `p & q`   | `p` E `q`                      |
| `^`     | `p ^ q`   | `p XOR` (ou exclusivo)         |

**Sempre** certifique-se de envolver cada condição dentro de `()` ao combinar máscaras booleanas com comparações bit a bit!


```python
elections[(elections['Year'] == 2008) | (elections['%'] >= 60)]
```

```python
elections[(elections['Year'] > 2000) & (elections['Result'] == 'win')]
```

```python
elections[(elections['Year'] < 2000) & (elections['Year'] > 1941) & (elections['Result'] == 'win') & (elections['%'] >= 55)]
```

```python
elections.loc[((elections['Party'] == 'Democratic') | (elections['Party'] == 'Republican')) & (elections['Year'] > 1900), ['Year', 'Party', '%']]
```

```python
third_party = elections[(elections['Party'] != 'Democratic') & (elections['Party'] != 'Republican') & (elections['Year'] > 1900)]
```
