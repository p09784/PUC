```python
import pandas
```

```python
dataset = pandas.read_csv('titanic.csv')
```

```python
dataset.info()
```

```python
dataset['age'] = dataset['age'].astype('str')
```

```python
dataset.info()
```

```python
dataset['age'] = dataset['age'].astype('float64')
```

```python
dataset.info()
```

```python
dataset['age'] = dataset['age'].fillna(dataset['age'].median())
```

```python
dataset['embarked'] = dataset['embarked'].fillna(dataset['embarked'].mode()[0])
```

```python
dataset.info()
```

```python
dataset.dropna(subset = ['boat'], inplace = True)
```
