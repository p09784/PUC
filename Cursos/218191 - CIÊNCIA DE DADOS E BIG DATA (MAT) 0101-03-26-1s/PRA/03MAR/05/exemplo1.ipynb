## Adicionando, removendo e modificando colunas

Para adicionar um coluna, use `[]` para referenciar a nova coluna desejada e, em seguida, atribua a ela uma `Series` ou um vetor com o comprimento apropriado.

```python
# crie uma Series com o comprimento de cada nome
babyname_lengths = babynames['Name'].str.len()

# adiciona um Series como um coluna chamada 'name_lengths'
babynames['name_lengths'] = babyname_lengths
```

Para modificar uma coluna, use `[]` - ou `.loc[]`- para acessar a coluna desejada e, em seguida, atribua-a novamente a um novo vetor ou Series.

```python
# modifique a coluna 'name_lengths' para que seja uma unidade menor do que seu valor original
babynames['name_lengths'] = babynames['name_lengths'] - 1
```

Para renomear uma coluna, use `.rename()`

```python
babynames = babynames.rename(columns = {'name_lengths' : 'Length'})
```

Para remover uma coluna, use `.drop`

```python
babynames = babynames.drop('Length', axis = 'columns')
```

## Ordenação

```python
babynames.sort_values('Count')
```

```python
babynames['Name'].sort_values()
```

#### Criar uma coluna temporária

```python
babyname_lengths = babynames['Name'].str.len()

babynames['name_lengths'] = babyname_lengths

babynames = babynames.sort_values(by = 'name_lengths', ascending = False)

babynames = babynames.drop('name_lengths', axis = 'columns')

babynames.head()
```

#### Ordenação usando o argumento `key`

```python
babynames.sort_values('Name', key = lambda name : name.str.len(), ascending = False).head()
```

### Ordenação usando a função `map`

```python
babynames['dr_ea_count'] = babynames['Name'].map(lambda name : name.count('dr') + name.count('ea'))

babynames.sort_values(by = 'dr_ea_count', ascending = False).head()
```
